```python
import duckdb
import pandas as pd
```

```python
# create a connection to the database
con = duckdb.connect('/workspaces/practice/terminology.db')
```

```python
con.execute('create schema if not exists METATHESAURUS')
```

```python
# use pandas to read the csv file
MRCONSO = pd.read_csv('/workspaces/practice/terminology/umls_metathesaurus/MRCONSO.RRF', sep='|', header=None, names=['CUI', 'LAT', 'TS', 'LUI', 'STT', 'SUI', 'ISPREF', 'AUI', 'SAUI', 'SCUI', 'SDUI', 'SAB', 'TTY', 'CODE', 'STR', 'SRL', 'SUPPRESS', 'CVF', 'RESERVE'], dtype={"SCUI": str, "SDUI": str, "CODE": str})

MRCONSO = MRCONSO.drop(columns=['RESERVE'])

con.execute('create or replace table METATHESAURUS.MRCONSO as select * from MRCONSO')

# show first 5 rows of RXATOMARCHIVE table
con.execute('select * from METATHESAURUS.MRCONSO limit 15').fetchdf()
```

```python
con.close()
```
